```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype='int')
dataset
```

```python
dataset.columns
```

```python
independent =dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['squared_error','absolute_error','friedman_mse','poisson'],'max_features':['auto','log2','sqrt'],'n_estimators':[10,100]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
result=grid.cv_results_
table=pd.DataFrame.from_dict(result)
table
```

```python
print('The r2_score value for the best parameters {}'.format(grid.best_params_)) 
```

```python
y_pred=grid.predict(independent)
from sklearn.metrics import r2_score
r_score=r2_score(dependent,y_pred)
r_score
```
