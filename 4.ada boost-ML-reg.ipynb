```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype='int')
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
param_grid={'loss':['linear','square','exponential'],'n_estimators':[50,100,150],'learning_rate':[0.01,0.1,1]}
grid=GridSearchCV(AdaBoostRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
result=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
print('The r2_score value for the best parameters {}'.format(grid.best_params_))
```

```python
y_pred=grid.predict(independent)
from sklearn.metrics import r2_score
r_score=r2_score(dependent,y_pred)
r_score
```
